The negative control sequencing from 3-3-22 also returned a number of 'valid' barcodes. This notebook further analyzes identity of 'valid' barcodes and degree of overlap with sequenced library samples.

```python
import pandas as pd
from plotnine import *
from dms_variants.constants import CBPALETTE
import dms_variants.plotnine_themes

from functools import reduce
```

```python
_ = theme_set(dms_variants.plotnine_themes.theme_graygrid())
```

```python
no_inf = (
    pd.read_csv("../../results/barcode_runs/counts_by_sample/libA_220303_1_no-antibody_control_AL049_no-infection.csv")
    .drop(['library', 'sample'], axis=1)
    .rename({'count': 'no-infection'}, axis=1)
)

no_inf = no_inf.drop(no_inf[no_inf['no-infection'] == 0].index)

no_rt = (
    pd.read_csv("../../results/barcode_runs/counts_by_sample/libA_220303_1_no-antibody_control_1_AL049_no-RT.csv")
    .drop(['library', 'sample'], axis=1)
    .rename({'count': 'no-rt'}, axis=1)
)

no_rt = no_rt.drop(no_rt[no_rt['no-rt'] == 0].index)
```

```python
no_rt
```

```python
libA_virus = (
    pd.read_csv("../../results/barcode_runs/counts_by_sample/libA_220224_1_no-antibody_control_670ng_1.csv")
    .drop(['library', 'sample'], axis=1)
    .rename({'count': 'libA-virus'}, axis=1)
)

libA_plasmid = (
    pd.read_csv("../../results/barcode_runs/counts_by_sample/libA_220224_1_no-antibody_control_plasmid.csv")
    .drop(['library','sample'], axis=1)
    .rename({'count': 'libA-plasmid'}, axis=1)
)

libA_pacbio = (
    pd.read_csv("../../results/barcode_runs/counts_by_sample/libA_220303_1_no-antibody_control_pacbio-prep.csv")
    .drop(['library','sample'], axis=1)
    .rename({'count': 'libA-pacbio-prep'}, axis=1)
)

libB_plasmid = (
    pd.read_csv("../../results/barcode_runs/counts_by_sample/libB_220303_1_no-antibody_control_plasmid.csv")
    .drop(['library','sample'], axis=1)
    .rename({'count': 'libB-plasmid'}, axis=1)
)

libB_pacbio = (
    pd.read_csv("../../results/barcode_runs/counts_by_sample/libB_220303_1_no-antibody_control_pacbio-prep.csv")
    .drop(['library','sample'], axis=1)
    .rename({'count': 'libB-pacbio-prep'}, axis=1)
)
```

```python
df_list = [no_inf, no_rt, libA_virus, libA_plasmid, libA_pacbio]

neg_ctrl_valid = reduce(lambda left,right: pd.merge(left, right, on=['barcode'], how='inner'), df_list)

neg_ctrl_valid
```

```python
neg_ctrl_valid_top = neg_ctrl_valid.drop(neg_ctrl_valid[(neg_ctrl_valid['no-infection']<1000) | 
                                                        (neg_ctrl_valid['libA-virus']<500)].index)

neg_ctrl_valid_top
```

```python
neg_ctrl_tidy = neg_ctrl_valid_top.melt(id_vars=['barcode'], 
                                             value_vars = ['libA-virus', 'libA-plasmid', 'libA-pacbio-prep', 'no-infection', 'no-rt'],
                                             var_name='sample',
                                             value_name='count'
                                            )

neg_ctrl_tidy
```

```python
p = (
    ggplot(
        neg_ctrl_tidy,
        aes(x='barcode', y='count', fill='sample')) +
    geom_point(size=3) + 
    theme(figure_size=(10, 5),
          panel_grid_major_x=element_blank(),
          axis_text_x=element_blank(),
         ) + 
    labs(title='60 valid barcodes at high freq in both no-infection and libA-virus') +
    scale_y_log10() +
    scale_fill_manual(values=CBPALETTE[1:])
)

_ = p.draw()

p.save('valid_bc_negative_ctrls.pdf')
```
