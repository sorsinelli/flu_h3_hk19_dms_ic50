```python
import pandas as pd
```

```python
def convert_df(file, ab):
    df = pd.read_csv(file)
    
    df = df.drop(columns=['epitope', 'n mutations', 'total negative'])
    df = df.rename(columns={'mean': 'site_mean', 
                            'total positive': 'site_Positive Differential Selection',
                            'max': 'mut_max', 
                            'min': 'mut_min'
                           })
    df['label_site'] = df['site']
    df['protein_site'] = df['site']
    df['mutation'] = 'F'
    df['protein_chain'] = 'A'
    
    df['condition'] = f'{ab}'
    
    df.to_csv(f'{ab}_dms-view.csv')
    
    return df
```

```python
ab_list = ['1c04', '5g04', 'cocktail']

for ab in ab_list:
    convert_df(f'{ab}_escape_df.csv', f'{ab}')
```

```python
ausab_07 = convert_df('ausab-07_escape_df.csv', 'AUSAB-07')
ausab_07
```
